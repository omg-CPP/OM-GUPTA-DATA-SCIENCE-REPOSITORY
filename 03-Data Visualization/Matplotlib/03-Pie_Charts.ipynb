## Matplotlib Pie Charts

```python
import matplotlib.pyplot as plt 
plt.style.use("ggplot")  # Choose any style you like
```

```python
# Basic Pie Chart Example
```

```python
# Data 
labels = ["Sachin", "Sehwag", "Kohli", "Yuvraj"] 
runs = [18000, 8000, 12000, 9500]
```

```python
plt.title("Career Runs of Indian Batsmen") 
plt.pie(runs, labels=labels) 
plt.show()
```

```python
# Custom Colors
# You can use color names or hex codes.
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99'] 
plt.pie(runs, labels=labels, colors=colors) 
plt.show()
```

- Add Edges and Style Slices with wedgeprops

```python
plt.pie( 
    runs, 
    labels=labels, 
    colors=colors, 
    wedgeprops={'edgecolor': 'black', 'linewidth': 2, 'linestyle': '--'} 
) 
plt.show() 
```

- Highlight a Slice Using explode

```python
explode = [0.1, 0.4, 0, 0]  # Only highlight Sachin's slice 
plt.pie(runs, labels=labels, explode=explode, colors=colors) 
plt.title("Exploded Pie Example") 
plt.show() 
```

- Add Shadows for a 3D Feel


```python
plt.pie( 
    runs, 
    labels=labels, 
    explode=explode, 
    colors=colors, 
    shadow=True  # adds a 3D-like shadow 
) 
plt.show() 
```

- Start Angle

```python
plt.pie( 
    runs, 
    labels=labels, 
    explode=explode, 
    colors=colors, 
    shadow=True, 
    startangle=50 
) 
plt.show() 
```

- Show Percentages with autopct

```python
plt.pie( 
    runs, 
    labels=labels, 
    autopct='%1.1f%%', 
    startangle=140 
) 
plt.title("Career Run Share") 
plt.show()
```

- Crowded Pie Chart – Why to Avoid

```python
# Too many categories 
languages = ['Python', 'Java', 'C++', 'JavaScript', 'C#', 'Ruby', 'Go', 'Rust', 
'Swift', 'PHP'] 
usage = [30, 20, 10, 10, 7, 5, 4, 3, 2, 1] 
plt.pie(usage, labels=languages, autopct='%1.1f%%', startangle=90) 
plt.title("Programming Language Usage (Crowded Example)") 
plt.show()
```

#### Why avoid it?
- Hard to compare slice sizes visually
- Cluttered and confusing
- No clear insight
Better alternatives: bar charts or horizontal bar charts

# assignment
- Create a pie chart showing the market share of mobile OS (Android, iOS, others). 
- Then recreate the same using a horizontal bar chart and observe which is easier to understand

```python
import matplotlib.pyplot as plt
```

```python
# Mobile operating system market share (in percentage)
os_names = ["Android", "iOS", "Others"]
market_share = [72, 27, 1]
```

```python
colors = ['#ff9999','#66b3ff','#99ff99']
plt.pie(
    market_share,
    labels=os_names,
    autopct="%1.1f%%",   # Show percentages
    startangle=90,       # Rotate chart so first slice starts at top
    explode = [0.2,0,0],
    colors = colors,
    shadow = True
)
plt.title("Mobile OS Market Share (Pie Chart)")
plt.show()
```

```python
# -------------------------------
# 2. Horizontal Bar Chart
# -------------------------------
plt.barh(os_names, market_share)
plt.xlabel("Market Share (%)")
plt.ylabel("OS")
plt.title("Mobile OS Market Share (Horizontal Bar Chart)")
plt.xlim(0, 100)  # Percentage scale from 0 to 100
plt.show()
```
