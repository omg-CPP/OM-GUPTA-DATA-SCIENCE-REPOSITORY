## Lambda Functions

#### 1. Square of a Number

```python
square = lambda x: x * x
print(square(5))
```

#### 2. Add Two Numbers

```python
add = lambda a, b: a + b
print(add(3, 4))
```

#### 3. Check Even or Odd

```python
is_even = lambda x: "Even" if x % 2 == 0 else "Odd"
print(is_even(10))
```

#### 4. Maximum of Two Numbers

```python
maximum = lambda a, b: a if a > b else b
print(maximum(8, 12))
```

#### 5. Sort a List of Tuples

```python
students = [("Om", 85), ("Rahul", 92), ("Amit", 78)]
students.sort(key=lambda x: x[1])
print(students)
```

```python
students = [("Om", 85), ("Rahul", 92), ("Amit", 78)]
students.sort(key=lambda x: x[1], reverse=True)
print(students)
```

#### 6. Using map()

```python
numbers = [1, 2, 3, 4]
squares = list(map(lambda x: x * x, numbers))
print(squares)
```

#### 7. Using filter()

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = list(filter(lambda x: x % 2 == 0, numbers))
print(evens)
```

```python
# using function
def is_even(x):
    return x % 2 == 0

evens_fun = list(filter(is_even, numbers))
print(evens_fun)
```

```python
# using list comprehension
evens_lc = [x for x in numbers if x % 2 == 0]
print(evens_lc)
```

## List Comprehension

#### 1. Create a List of Squares

```python
numbers = [1, 2, 3, 4, 5]
squares = [x * x for x in numbers]
print(squares)
```

#### 2. Filter Even Numbers

```python
numbers = [1, 2, 3, 4, 5, 6]
evens = [x for x in numbers if x % 2 == 0]
print(evens)
```

#### 3. Filter Odd Numbers

```python
numbers = [1, 2, 3, 4, 5, 6]
odds = [x for x in numbers if x % 2 != 0]
print(odds)
```

#### 4. Convert Strings to Uppercase

```python
names = ["om", "rahul", "amit"]
upper_names = [name.upper() for name in names]
print(upper_names)
```

#### 5. Conditional Expression (Adult/Minor)

```python
ages = [12, 18, 25, 15]
status = ["Adult" if age >= 18 else "Minor" for age in ages]
print(status)
```

#### 6. Extract First Letter of Each Word

```python
words = ["apple", "banana", "cherry"]
first_letters = [word[0] for word in words]
print(first_letters)
```

#### 7. Flatten a Nested List

```python
matrix = [[1, 2], [3, 4], [5, 6]]
flat = [num for row in matrix for num in row]
print(flat)
```

#### 8. Generate Pairs

```python
pairs = [(x, y) for x in range(3) for y in range(2)]
print(pairs)
```

#### 9. Tuples in Zip

```python
l1 = [1,2,3,4,5]
l2 = [1,2,3,4,5]
tuples = [(x,y) for x,y in zip(l1,l2)]
print(tuples)
tuples_list = [x+y for x,y in zip(l1,l2)]
print(tuples_list)
```
