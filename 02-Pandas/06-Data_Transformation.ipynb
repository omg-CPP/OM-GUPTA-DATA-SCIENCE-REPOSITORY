## Data Transformation
- Once your data is clean, the next step is to reshape, reformat, and reorder it as needed for analysis. 
- Pandas gives you plenty of flexible tools to do this.

```python
import pandas as pd
```

```python
# Sorting & Ranking
```

```python
df = pd.read_csv("data2.csv")
```

```python
df
```

```python
df.sort_values("Year")
```

```python
df.sort_values("Box Office")
```

```python
df.sort_values("Year", ascending=False) # Descending
```

```python
df.sort_values(["IMDB", "Year"]) # Sort by multiple columns
```

```python
df2 = df.sort_values(["IMDB", "Year"]).copy()
df2
```

```python
# Reset the index and drop the old index
df2.reset_index(drop=True, inplace=True)
```

```python
df2
```

```python
df3 = df.sort_values(["IMDB", "Year"]).copy()
```

```python
df3
```

```python
df3.sort_index()
```

```python
df2
```

```python
df2["Rank"] = df2["IMDB"].rank(ascending=False, method="dense")
```
