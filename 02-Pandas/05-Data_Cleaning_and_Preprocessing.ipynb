## Data Cleaning and Pre-Processing

```python
df
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.dropna() # drops all rows with NULL values
```

```python
df.bfill() # Backward fill
```

### Applying Functions

```python
# .apply() → Apply any function to rows or columns
```

```python
df2
```

```python
# we could build a function also
def isMinor(x):
    return "Adult" if x >= 25 else "Minor"
df2["Age Group"] = df2["Age"].apply(lambda x: "Adult" if x >= 25 else "Minor")
```

```python
df2
```

```python
# .map() → Element-wise mapping for Series
```

```python
gender_map = {"M": "Male", "F": "Female"}
df2["Gender"] = df2["Gender"].map(gender_map)
```

```python
df2
```

```python
# .replace() → Replace specific values
```

```python
city_map = {"New York": "New Delhi", "Mumbai": "Navi Mumbai"}
df2["City"] = df2["City"].replace(city_map)
```
