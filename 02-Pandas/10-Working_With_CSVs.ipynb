```python
# DataFrame 1: Student scores
df1 = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Math": [85, 78, 92],
    "Science": [90, 88, 95]
})

# DataFrame 2: Employee information
df2 = pd.DataFrame({
    "Employee": ["John", "Emma", "Raj"],
    "Department": ["IT", "HR", "Finance"],
    "Salary": [60000, 55000, 70000]
})

print(df1)
print("\n")
print(df2)
```

- for writing to multiple sheets

```python
with pd.ExcelWriter("report.xlsx") as writer:
    df1.to_excel(writer, sheet_name="Student", index=False)
    df2.to_excel(writer, sheet_name="Employee", index=False)
```
