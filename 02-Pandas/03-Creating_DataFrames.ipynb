## Creating DataFrames
- Let’s look at different ways to create a Pandas 
DataFrame — the core data
structure you’ll be using 90% of the time in data science.

```python
import pandas as pd
```

- From Python Lists

```python
data = [['harry',34], ['jack',67], ['gill',23], ['ajay',56]]
```

```python
data
```

```python
df = pd.DataFrame(data, columns=['Name', 'Marks'])
```

```python
df
```

- From Dictionary of Lists

```python
data = {
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [25, 30, 35]
}
df = pd.DataFrame(data)
df
```

```python
df.index = ['1','2','3']
```

```python
df
```

```python
# Each key becomes a column, and each list is the column data.
```

- From NumPy Arrays

```python
import numpy as np
```

```python
arr = np.array([[1,2], [5,6]])
```

```python
df = pd.DataFrame(arr, columns=['A','B'])
```

```python
df
```

- From CSV Files

```python
df = pd.read_excel("data.xlsx")
df
```

```python
df = pd.read_csv("data.csv")
df
```

- JSON file

```python
df = pd.read_json("data.json")
```

```python
df
```

- From URL

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)
```

```python
df
```

## EDA - Exploratory Data Analysis

- Exploratory Data Analysis (EDA) is an essential first step in any data science project.
- It involves taking a deep look at the dataset to understand its structure, spot
patterns, identify anomalies, and uncover relationships between variables. 
- This process includes generating summary statistics, checking for missing or duplicate
data, and creating visualizations like histograms, box plots, and scatter plots. The
goal of EDA is to get a clear picture of what the data is telling you before applying
any analysis or machine learning models.
