Functions in Python

```python
# defining a function
def greet(name):
    return f"Hello, {name}!"
```

```python
# calling a function
print(greet("Alice"))
```

```python
# function with default argument
def greet(name="Guest"):
    return f"Hello, {name}!"
```

```python
greet()
```

```python
# function with multiple arguments
def add(a,b):
    return a + b;
print(add(3,5))
```

```python
def avg(a,b):
    return (a + b) / 2
```

```python
avg(4,6)
```

```python
# default arguments always come at the end
def sum(a,b,c=10):
    return a + b + c
```

```python
sum(3,4)
```
