With

```python
with open("text.txt", "r") as file:
    content = file.read()
print(content) # No need to manually close the file
```

```python
with open("harry.txt", "r") as file:
    content = file.read()
print(content) # No need to manually close the file
```

```python
with open("DS.txt", "r") as file:
    content = file.read()
print(content) # No need to manually close the file
```
