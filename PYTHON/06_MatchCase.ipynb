```python
def http_status(code):
    match code:
        case 200:
            return "OK"
        case 400:
            return "Bad Request"
        case 404:
            return "Not Found"
        case 500:
            return "Internal Server Error"
        case _:
            return "Unknown Status"
print(http_status(200)) # Output: OK
print(http_status(404)) # Output: Not Found
```

```python
print(http_status(500))
```

```python
print(http_status(700))
```
