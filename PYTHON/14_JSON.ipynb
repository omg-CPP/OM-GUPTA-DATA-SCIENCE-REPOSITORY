JSON (JavaScript Object Notation) is a lightweight data format used for data
exchange between servers and applications. It is widely used in APIs, web
applications, and configurations.

Python provides the 
json module to work with JSON data. You can import the
json module like this: import json

```python
import json
```

Converting Python Objects to JSON (Serialization)

```python
obj = {"name": "Alice", "company": "IBM", "salary": 45000}
```

```python
s = json.dumps(obj)
print(type(s))
s
```

```python
# Converting JSON to Python Objects (Deserialization)
d = json.loads('{"name": "Alice", "company": "IBM", "salary": 45000}')
```

```python
print(type(d))
d
```

json.dump() – Write JSON data to a file


```python
file = open("data.json", "w")
```

```python
json.dump(obj, file)
```

```python
file.close()
```

```python
type(obj)
```

```python
with open("data.json", "r") as file:
    python_data = json.load(file)
print(python_data) 
```

```python
type(python_data)
```

Formatting JSON Output

```python
formatted_json = json.dumps(obj, indent=4)
print(formatted_json)
```

```python
type(obj)
```
