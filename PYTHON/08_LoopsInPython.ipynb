For Loop

```python
l = [3,4,5,6,7]
```

```python
for item in l:
    print(item)
```

While Loop

```python
count = 0
while(count < 3):
    print(count)
    count += 1
```

```python
i = 0
while(i < 12):
    print(i)
    i += 1
```

Loop Control Statements

```python
for item in l:
    pass # don't do anything
print("hello")
```

```python
list = ["apples", "bananas", "tomatoes"]
```

```python
for item in list:
    print(item)
```

```python
for item in list:
    if(item == "bananas"):
        break # stops this loop right here
    print(item)
```

```python
for item in list:
    if(item == "bananas"):
        continue # skips this particular iteration of the loop
    print(item)
```

```python
for i in range(5):
    if i == 3:
        break # Stops the loop at 3
    print(i)
```
