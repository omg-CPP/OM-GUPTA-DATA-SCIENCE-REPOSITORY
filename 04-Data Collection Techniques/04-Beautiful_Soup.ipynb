- 1. What is BeautifulSoup?
  - BeautifulSoup is a Python library used to parse HTML and XML documents.
  - It creates a parse tree from page content, making it easy to extract data.
  - It is often used with requests to scrape websites.

### Navigating HTML Structure in Scraping
#### Scraping tools like BeautifulSoup and Selenium use tag names and attributes to locate elements.
Example Targets
- By tag: soup.find('div')
- By class: soup.find('div', class_='product')
- By ID: soup.find(id='header')
- By attribute: soup.find('a', {'href': True})

```python
articles = soup.find_all("article", class_="product_pod")
articles
```

```python
articles = soup.select("article.product_pod")
articles
```

```python
# using html
for article in articles:
    title = article.find("h3").find("a")["title"]
    price = article.find("div", class_="product_price").find("p", class_="price_color").text
    print(title)
    print(price)
```

```python
# using css
for article in articles:
    title = article.find("h3").find("a")["title"]
    price = article.select_one("div.product_price p.price_color").text
    print(title)
    print(price)
```

```python
#.find() vs .select_one()
#   Method	        Syntax	              Example
#- .find()	    Tag + attributes	find("div", class_="product_price")
#- .select_one()	CSS selector	    select_one("div.product_price")
```

```python
items = []
for article in articles:
    title = article.find("h3").find("a")["title"]
    price = article.select_one("p.price_color").text.split("£")[1]
    rating_element = article.select_one("p.star-rating")
    rating = rating_element['class'][1]
    print(rating)
    items.append([title, price, rating])
```

```python
items
```

```python
import pandas as pd
```

```python
# creating dataframe from list of list
df = pd.DataFrame(items, columns=["Book Title", "Price", "Rating"])
df
```
